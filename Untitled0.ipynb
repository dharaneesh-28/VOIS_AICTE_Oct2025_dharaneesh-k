```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
data = {
    "neighbourhood group": ["A","A","B","B","C","C","D","D","E","E"],
    "price": [100,200,150,300,250,400,120,220,180,350],
    "construction year": [2001,2005,2010,2015,2020,2008,2003,2011,2016,2019],
    "service fee": [10,20,15,30,25,40,12,22,18,35],
    "review rate number": [4,5,3,4,5,2,5,3,4,4],
    "host name": ["John","Mary","Sam","John","Mary","Alex","Sam","Mary","John","Alex"],
    "calculated host listings count": [2,5,3,10,7,12,6,4,8,9],
    "availability 365": [180,200,150,365,300,250,100,320,280,360],
    "host_identity_verified": ["t","f","t","t","f","t","f","t","t","f"],
    "room type": ["Private room","Entire home","Shared room","Private room","Entire home",
                  "Private room","Entire home","Shared room","Private room","Entire home"]
}
df = pd.DataFrame(data)

counts = df.groupby('neighbourhood group').size().sort_values(ascending=False)
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)

print("Listings by Neighbourhood Group:\n", counts, "\n")
print("Average Price by Neighbourhood Group:\n", avg_price, "\n")

plt.figure(figsize=(7,4))
sns.barplot(x=counts.index, y=counts.values, palette='Blues_d')
plt.title('Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.show()

plt.figure(figsize=(7,4))
sns.barplot(x=avg_price.index, y=avg_price.values, palette='Greens_d')
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='construction year', y='price', alpha=0.6)
sns.regplot(data=df, x='construction year', y='price', scatter=False, color='red')
plt.title('Price vs Construction Year')
plt.show()

host_counts = df.groupby('host name')['calculated host listings count'].max().sort_values(ascending=False).head(10)
print("Top 10 Hosts by Listings:\n", host_counts, "\n")

plt.figure(figsize=(7,4))
sns.barplot(x=host_counts.values, y=host_counts.index, palette='mako')
plt.title('Top 10 Hosts by Listings Count')
plt.xlabel('Listings Count')
plt.ylabel('Host Name')
plt.show()

df['positive_review'] = np.where(df['review rate number'] >= 4, 1, 0)
id_rate = df.groupby('host_identity_verified')['positive_review'].mean()
print("Positive Review Rate by Verification:\n", id_rate, "\n")

plt.figure(figsize=(5,4))
sns.barplot(x=id_rate.index.astype(str), y=id_rate.values, palette='Set2')
plt.title('Positive Reviews by Verification')
plt.ylim(0,1)
plt.show()

corr_val = df[['price','service fee']].corr().iloc[0,1]
print("Correlation between Price and Service Fee:", round(corr_val,3), "\n")

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='price', y='service fee', alpha=0.6)
sns.regplot(data=df, x='price', y='service fee', scatter=False, color='red')
plt.title(f'Service Fee vs Price (corr = {round(corr_val,3)})')
plt.show()

pivot_rt = df.pivot_table(index='neighbourhood group', columns='room type',
                          values='review rate number', aggfunc='mean')
print("Pivot Table of Review Rates:\n", pivot_rt, "\n")

plt.figure(figsize=(7,4))
sns.heatmap(pivot_rt, annot=True, fmt='.2f', cmap='YlGnBu')
plt.title('Avg Review Rate by Neighbourhood & Room Type')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='calculated host listings count', y='availability 365', alpha=0.6)
sns.regplot(data=df, x='calculated host listings count', y='availability 365', scatter=False, color='red')
plt.title('Availability vs Host Listings Count')
plt.show()
```
